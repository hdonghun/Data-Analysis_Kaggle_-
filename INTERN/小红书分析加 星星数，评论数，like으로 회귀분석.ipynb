```python
data_name = f"NEW(20220210).csv"
df = pd.read_csv(data_name) #Encoding cp949는 못함
df
```

```python
# Nan 값 0으로 만들어주기
df = df.fillna(value=0)
df
```

```python
df = df.drop(["like.1"], axis=1)
df = df.drop(["like.2"], axis=1)
df = df.drop(["星星数.1"], axis=1)
df = df.drop(["星星数.2"], axis=1)
df = df.drop(["评论数.1"], axis=1)
df = df.drop(["评论数.2"], axis=1)
df = df.drop(["序号"], axis=1)
```

```python
ndf = df[(df['星星数'] != 0.0) & (df['评论数'] != 0.0) & (df['like'] != 0.0)].sort_values(by=['星星数'],ascending=False)
ndf
```

```python
ndf.plot(kind='scatter', x='赞藏总数', y='星星数', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='评论数', y='赞藏总数', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='like', y='赞藏总数', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='星星数', y='粉丝数', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='评论数', y='粉丝数', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='like', y='粉丝数', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='星星数', y='价钱', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='评论数', y='价钱', c='coral', s=10, figsize=(5,3))
ndf.plot(kind='scatter', x='like', y='价钱', c='coral', s=10, figsize=(5,3))


plt.show()
plt.close()
```

```python
# Seaborn 라이브러리의 regplot() 함수를 이용하여 두 변수에 대한 그래프(산점도) 그리기.

fig = plt.figure(figsize=(10,5))
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
sns.regplot(x='评论数', y='赞藏总数', data=ndf, ax=ax1)  # 회귀선 표시
sns.regplot(x='星星数', y='赞藏总数', data=ndf, ax=ax2) # 회귀선 표시 안함 표시는 , fit_reg = False
    
plt.show()
plt.close()
```

```python
grid_ndf = sns.pairplot(ndf)
plt.show()
plt.close()
```

```python
X=ndf[['星星数']] # 독립 변수 X
y=ndf[['like']]    # 종속 변수 Y

# train data와 test data로 구분(7:3 비율)
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,            # 독립 변수
                                                   y,             # 종속 변수
                                                   test_size=0.3,  # 검증 30%
                                                   random_state=10 # 랜덤 추출 값
                                                   )

print('train data 개수 : ', len(X_train))
print('testdata 개수 : ', len(X_test))
```

```python
# sklearn 라이브러리에서 선형회귀분석 모듈 가져오기
from sklearn.linear_model import LinearRegression

# 단순 회귀 분석 모형 객체 생성
lr = LinearRegression()

# train data를 가지고 모형 학습
lr.fit(X_train, y_train)

# 학습을 마친 모형에 test data를 적용하여 결정계숙(R-제곱) 계산
r_square = lr.score(X_test, y_test)

print(r_square)
```

```python
# 회귀식의 기울기
print('기울기 : ', lr.coef_)
print('\n')

# 회귀식의 y절편
print('y절편 :', lr.intercept_)
```

```python
# 모형에 전체 X 데이터를 입력하여 예측한 값 y_hat을 실제 값 y와 비교 
y_hat = lr.predict(X)

plt.figure(figsize=(8, 5))
ax1 = sns.distplot(y, hist=False, color = 'green',label="y")

ax2 = sns.distplot(y_hat, hist=False, label="y_hat", color ='red', ax=ax1)
plt.show()
plt.close()
```
