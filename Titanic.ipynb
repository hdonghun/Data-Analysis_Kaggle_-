 * 불린 인덱싱(Boolean indexing)
 
 헷갈리는 위치기반, 명칭기반 인덱싱을 사용할 필요없이 조건식을 [ ] 안에 기입하여 간편하게 필터링을 수행.

```python
titanic_boolean = titanic_df[titanic_df['Age']>75]
print(type(titanic_boolean))
titanic_boolean
```

```python
titanic_df[titanic_df['Age']>60][['Name','Age']].head(3)
```

```python
titanic_df[['Name','Age']][titanic_df['Age']> 60].head(3)
```

위 두개의 방법 모두 같은 결과를 보여준다.

논리 연산자로 결합된 조건식도 불린 인덱싱으로 적용 가능합니다.

```python
titanic_df[ (titanic_df['Age'] > 60) & (titanic_df['Sex']=='female') & (titanic_df['Pclass']==1)]
```

위 처럼 조건식이 길어지면 가독성이 떨어지기 때문에, 아래와 같은 방법으로 한다.!

```python
x1 = titanic_df['Age'] > 60
x2 = titanic_df['Pclass'] == 1
x3 = titanic_df['Sex'] =='female'
titanic_df[x1&x2&x3]
```
