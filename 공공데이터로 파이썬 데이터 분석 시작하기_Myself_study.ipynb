```python
#전용면적이라는 반복되는 이름을 제거하여 데이터를 값을 줄이는 과정
df_last["규모구분"].unique()
```

```python
#전용면적이라는 반복되는 이름을 제거하여 데이터를 값을 줄이는 과정
#
df_last["전용면적"] = df_last["규모구분"].str.replace("전용면적","")
df_last["전용면적"] = df_last["전용면적"].str.replace("초과","~")
df_last["전용면적"] = df_last["전용면적"].str.replace("이하","")
df_last["전용면적"] = df_last["전용면적"].str.replace(" ","")
df_last["전용면적"]
```

```python
#사용하지 않는 컬럼 제거하기 #메모리 사용량을 줄일 수 있다.
df_last = df_last.drop(["규모구분","분양가격(제곱미터)"],axis=1)
df_last
```

#groupby 로 데이터 집계하기
groupby를 통해 데이터를 그룹화해서 연산하기

#지역명으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 하기.

#df.groupby(["인덱스로 사용할 컬럼명"])["계산할 컬럼 값"].연산()


```python
df_last.groupby(["지역명"])["평단분양가격"].mean() #평균값
```

```python
#전용면적으로 분양가격의 평균을 구합니다.
df_last.groupby(["전용면적"])["평단분양가격"].mean()
```

```python
# 두 가지의 인데스로 평당분양가격의 평균 구하기
df_last.groupby(["전용면적","지역명"])["평단분양가격"].mean()
#unstack() : 끝에 있는 인덱스가 columns로 와서 더 이쁘게 보여준다.
#.round() : 뒤에 지저분한 소수점이 한자리수로 정리되어 더 이쁘게 보여준다.
```

```python
df_last.groupby(["전용면적","지역명"])["평단분양가격"].mean().unstack().round()
```

```python
#연도, 지역명으로 평단분양가격의 평균을 구합니다.
g = df_last.groupby(["연도", "지역명"])["평당분양가격"].mean().unstack().T
g
```

pivot table 로 위에서 한 groupby로 한 작업 똑같이 해보기

```python
pd.pivot_table(df_last, index=["지역명"], values=["평단분양가격"], aggfunc="mean")
#groupby 와 pivot table의 차이는 연산의 결과가 
#series로 나오는냐 Dataframe형식으로 나오느냐 차이다.
```

```python
#둘의 차이점 보기
#df_last.groupby(["전용면적"])["평단분양가격"].mean()  >>series 형태
pd.pivot_table(df_last, index="전용면적", values="평당분양가격")
```

```python
#둘이 비교
#df_last.groupby(["전용면적","지역명"])["평단분양가격"].mean().unstack().round()
df_last.pivot_table(index="전용면적", columns = "지역명", values="평당분양가격" ).round()
```

```python
#연도, 지역명으로 평단분양가격의 평균을 구합니다.
g = df_last.groupby(["연도", "지역명"])["평당분양가격"].mean().unstack().T
p = df_last.pivot_table(index=["연도","지역명"], values="평당분양가격")
p
```

최근 데이터 시각화 하기 

데이터 시각화를 우한 폰트 설정

(한글 폰트 사용을 위해 matplotlib의 pyplot을 plt라는 별칭으로 불러오기)

```python
import matplotlib.pyplot as plt
plt.rc("font", family="Malgun Gothic")
```

```python
#지역명으로 분양가격의 평균을 구하고 선그래프로 시각화하기
g = df_last.groupby(["지역명"])["평당분양가격"].mean().sort_values(ascending = False)
g.plot()
```

```python
g.plot.bar(rot=0, figsize=(10,3))
```

```python
#전용면적으로 분양가격의 평균을 구하고 막대그래프(bar)로 시각화 합니다.
df_last.groupby(["전용면적"])["평당분양가격"].mean().plot.bar()
```

```python
#연도별 분양가격의 평균을 구하고 막대 그래프로 시각화
df_last.groupby(["연도"])["평당분양가격"].mean().plot()
```
